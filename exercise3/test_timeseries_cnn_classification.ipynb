```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading csv from file
df = pd.read_csv("walmart_cleaned.csv",index_col=0)
df.head()
```

```python
df.describe()
```

```python
# Taking a look at how the stores are represented in the data
print(df['Store'].value_counts())

# Taking a look at the data of a random store, I chose store 13
df_store = df[df['Store'] == 13]
df_store.describe()
```

```python
# Printing the amount of how often each department from the store is represented in the data
value_count_list = df_store["Dept"].value_counts().tolist()
print(value_count_list)

# Seems like most departments exist 143 times in the data for that store
# Let me see if that matches the number of weeks in the data

# There also are 143 weeks of data
# So I will sum the data for each store together to get 143 weeks * 45 stores
# => 6435 rows of data in total
print(df_store["Date"].nunique()) 
print(df["Store"].nunique()) 
```

```python
# Combine weekly data by store number
# Average value for: fuel price, temperature a
# Sum for: weekly sales
# Holiday, unemployment rate, type, size, CPI should be identical for all rows for a given week/store
# I will drop the MarkDown columns, because they are determined by the store
# Taking them into account and achieve good results would probably be quite challenging
# The idea is to classify the demand/sales in categories based on relevant input variables
# e.g. weekly sales in the top 20% would be high demand, bottom 20% would be low demand
```

```python
def most_common_column_value(series):
    return series.value_counts().idxmax()
```

```python
print(df.groupby(['Date','Store']).size())

# Trying to combine the rows so I get weekly data by store, but for all departments together
# First dropping rows I don't want, including the index, mark downs
df.drop(['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5'], axis=1, inplace=True)
```

```python
df
```

```python
# Aggregating the rows, treating columns based on the values inside
aggregation_functions = {'IsHoliday': most_common_column_value, 'Weekly_Sales': 'sum', 'Temperature': 'mean', 'Fuel_Price': 'mean', 'CPI': 'mean', 'Unemployment': 'mean', 'Type': most_common_column_value, 'Size': most_common_column_value}
df_new = df.groupby(['Date', 'Store']).aggregate(aggregation_functions)
df_new = df_new.reset_index()
```

```python
# Plotting the weekly sales by store
# This is a bit of a rainbowy mess
# the stores clearly have some slight differences in sales
# but the overall trends are pretty close
fig, ax = plt.subplots(1, 1, figsize=(15, 4))

for i in range(45):
    df_store = df_new[df_new['Store'] == i+1]
    ax.plot(df_store["Date"], df_store["Fuel_Price"], lw=1, label='Weekly Sales')
plt.show()
```

```python
# Aggregating the rows, treating columns based on the values inside
aggregation_functions = {'IsHoliday': most_common_column_value, 'Weekly_Sales': 'sum', 'Temperature': 'mean', 'Fuel_Price': 'mean', 'CPI': 'mean', 'Unemployment': 'mean'}
df_all_stores = df.groupby(['Date']).aggregate(aggregation_functions)
df_all_stores = df_all_stores.reset_index()
```
